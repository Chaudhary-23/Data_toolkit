Q2- How does broadcasting work in NumPy?
Ans- Broadcasting in NumPy is a powerful feature that allows operations on arrays of different shapes without making multiple copies of data.
What is Broadcasting-
Broadcasting lets NumPy automatically expand the smaller array so it matches the shape of the larger array when performing element-wise operations like addition, multiplication, etc.

#Example:

import numpy as np

a = np.array([1, 2, 3])      # Shape: (3,)
b = np.array([[10], [20], [30]])  # Shape: (3, 1)
result = a + b

#Output -
[[11, 12, 13],
 [21, 22, 23],
 [31, 32, 33]]

Broadcasting Rules (simplified):
Compare shapes from the end (right to left).
Dimensions are compatible if:
They are equal, or
One of them is 1.
If a dimension is 1, it's stretched to match the other.

Q6- What are the differences between NumPy arrays and Python lists?
Ans-  While NumPy arrays and Python lists might look similar on the surface, they’re quite different under the hood—especially when it comes to performance and functionality.
Key differences:
Feature | Python List | NumPy Array
Data Type | Can hold mixed types (e.g., int, str, float) | Must hold a single data type (e.g., all floats or all ints)
Performance | Slower for numerical operations | Much faster due to C-level optimization
Memory Usage | Higher (stores extra info for each element) | Lower (tightly packed, fixed-type)
Operations | Element-wise operations need loops | Supports fast vectorized operations
Functionality | Basic, no built-in math tools | Comes with a huge set of numerical and linear algebra tools
Multidimensional Support | Awkward and limited (lists of lists) | Native support for multi-D arrays

#Example:
# Python list
lst = [1, 2, 3]
print([x * 2 for x in lst])  # Output: [2, 4, 6]

# NumPy array
import numpy as np
arr = np.array([1, 2, 3])
print(arr * 2)  # Output: [2 4 6]

Same result, but the NumPy version is cleaner, faster, and scales better.

Q8- What does the term “vectorized operation” mean in NumPy?
Ans-  A vectorized operation in NumPy refers to performing operations on entire arrays (or vectors) without using explicit loops—the operation is applied element-wise automatically and efficiently under the hood.

Instead of looping through each element manually, NumPy uses optimized C-based routines to do the job much faster.
#Example:
Without vectorization (pure Python)
a = [1, 2, 3]
b = [4, 5, 6]
result = [a[i] + b[i] for i in range(len(a))]

With vectorization (NumPy):
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a + b  # vectorized!

#Output: Both give [5, 7, 9], but the NumPy version is shorter, cleaner, and much faster, especially for large arrays.

Q10- What is the significance of hierarchical indexing in Pandas?
Ans- Hierarchical indexing in Pandas (also known as MultiIndex) allows you to work with multi-dimensional data in a 2D structure like a DataFrame or Series. It enables you to index and slice data with more than one level of indexing, which is especially useful when you have complex datasets with multiple layers of information.
Why is Hierarchical Indexing Important?
Organizing complex datasets: It’s useful when you have data that naturally falls into hierarchical levels. For example, data grouped by year, month, and day, or by region, city, and store.
Efficient data manipulation: Allows easy access, reshaping, and aggregation of data across different levels.
GroupBy operations: Facilitates grouping, aggregation, and transformation of data by hierarchical levels.
Multi-dimensional operations: Makes it easier to represent data in a more compact, yet powerful, way.

Creating a MultiIndex DataFrame:
#Example:
import pandas as pd
import numpy as np

# Create some example data
index = pd.MultiIndex.from_tuples(
    [('A', 'x'), ('A', 'y'), ('B', 'x'), ('B', 'y')],
    names=['Category', 'Subcategory']
)

data = np.random.randn(4, 2)
df = pd.DataFrame(data, index=index, columns=['Value1', 'Value2'])

print(df)

#Output:
                  Value1   Value2
Category Subcategory                
A        x            0.1216   -1.3234
         y           -0.2837    0.1342
B        x           -0.1575   -0.0672
         y            0.3982   -0.2923

In this example:
The index has two levels: Category and Subcategory.
You can easily access data at different levels using the .loc[] method.

Accessing Data with Hierarchical Indexing:
You can access different levels of the index by specifying them:

# Access data for category 'A' and subcategory 'x'
print(df.loc['A', 'x'])

#Output:
Value1    0.1216
Value2   -1.3234
Name: (A, x), dtype: float64

Advanced Use Cases:
Pivot tables: Automatically create multi-level indices when working with pivot tables.
Reshaping: Use .stack() and .unstack() to reshape data.
Aggregation: Perform aggregation operations across different levels of the index (e.g., groupby()).

Summary:
Hierarchical indexing is a powerful feature in Pandas for working with multi-dimensional data in a 2D structure.
It helps you organize, access, and manipulate complex data more intuitively.
It's especially useful for tasks like grouping, aggregating, and pivoting data.

Q13- Why is handling missing data important in Pandas?
Ans- Handling missing data is a crucial aspect of data cleaning and preprocessing because missing values can significantly affect the accuracy, integrity, and quality of your analysis. In Pandas, dealing with missing data properly ensures that your data analysis doesn't lead to biased, incorrect, or misleading results.

Why is Handling Missing Data Important?
Accuracy of Analysis:
If you leave missing data unaddressed, it may skew your analysis and lead to incorrect conclusions.
Many statistical and machine learning models cannot handle missing values and will produce errors or suboptimal results if missing data is present.

Bias in Results:
If missing data is ignored, it can introduce bias—especially if missing values are not randomly distributed but instead follow some underlying pattern (e.g., certain values missing for specific groups).

Compatibility with Algorithms:
Some machine learning algorithms and statistical tests expect complete data. Missing values may need to be handled or imputed (replaced) for those algorithms to work properly.

Data Integrity:
Ensuring that all of your data is consistent and properly filled or imputed improves the overall integrity of your dataset, which is key to effective decision-making.

Common Ways to Handle Missing Data in Pandas:
Identifying Missing Data:
Pandas has NaN (Not a Number) to represent missing data.
You can check for missing data using:

df.isnull()  # Returns a DataFrame of booleans
df.isna()  # Equivalent to isnull()

Removing Missing Data:
Remove rows with missing values using .dropna():

df.dropna()  # Removes any row containing NaN values

Remove columns with missing values:
df.dropna(axis=1)  # Removes columns with NaN values

When to Use: Removing missing data is best when the amount of missing data is small and not likely to introduce bias.

Filling Missing Data:
Fill missing values with a constant value:

df.fillna(0)  # Replace NaN with 0

Fill using forward/backward fill (carry the last valid value forward):
df.fillna(method='ffill')  # Forward fill
df.fillna(method='bfill')  # Backward fill

Fill with the mean, median, or mode:
df.fillna(df.mean())  # Fill NaN with the column mean

When to Use: Filling is useful when you believe the missing data can be reasonably estimated from other available data. This is common when missing values are random.

Imputation with Advanced Techniques:
For more sophisticated filling, such as using machine learning models or interpolation, you can use libraries like Scikit-learn or Statsmodels.

Replacing with Interpolation:
Interpolation is used to estimate missing values based on other data points:
df.interpolate()  # Linearly interpolates missing values

When to Use: Interpolation is great when you expect the missing values to follow a pattern or trend in time series data.


Example: Handling Missing Data:

import pandas as pd

# Create a sample DataFrame with missing values
data = {'A': [1, 2, None, 4], 'B': [None, 5, 6, 7], 'C': [8, None, 10, 11]}
df = pd.DataFrame(data)

# Check for missing data
print(df.isnull())

# Fill missing values with the column mean
df_filled = df.fillna(df.mean())

print(df_filled)

Summary:
Handling missing data is crucial because:
It ensures accurate analysis and avoids bias.
It prevents errors in statistical models and machine learning algorithms.
It allows you to make informed decisions when replacing or removing missing data.

Q15- How does NumPy handle multidimensional arrays?
Ans- NumPy is extremely efficient at handling multidimensional arrays, which are arrays with more than one axis (dimension). These are often referred to as ndarrays (N-dimensional arrays), and they are one of the core features of NumPy. Handling multidimensional arrays allows for complex data structures and operations, such as matrices or higher-dimensional datasets.

Here's a breakdown of how NumPy handles multidimensional arrays:

1. Creating Multidimensional Arrays:
You can create multidimensional arrays by passing a nested list (or a list of lists) to the np.array() function.

#Example:
import numpy as np

# 2D array (matrix)
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr_2d)

#Output:
[[1 2 3]
 [4 5 6]]

In this example, the array arr_2d is a 2D array (a matrix) with 2 rows and 3 columns.


More Dimensions:

# 3D array
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(arr_3d)

#Output:
[[[1 2]
  [3 4]]

 [[5 6]
  [7 8]]]

Here, arr_3d is a 3D array, representing a cube of values, with dimensions (2, 2, 2).


2. Shape and Dimensions of Arrays:
NumPy allows you to easily check the shape and dimensions of an array.
Shape gives the dimensions of the array (number of rows, columns, etc.).
ndim provides the number of dimensions of the array.

#Example:
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(arr_2d.shape)  # Output: (2, 3) - 2 rows, 3 columns
print(arr_2d.ndim)   # Output: 2 - 2 dimensions

print(arr_3d.shape)  # Output: (2, 2, 2) - 2 blocks, 2 rows, 2 columns
print(arr_3d.ndim)   # Output: 3 - 3 dimensions



3. Accessing Elements in Multidimensional Arrays:
You can access elements of a multidimensional array using indexing and slicing. For multidimensional arrays, you use a comma to separate indices for each dimension.

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Accessing specific elements
print(arr_2d[0, 1])  # Output: 2 (1st row, 2nd column)
print(arr_2d[1, 2])  # Output: 6 (2nd row, 3rd column)

# Slicing a 2D array
print(arr_2d[:, 1])  # Output: [2 5] (all rows, 2nd column)
print(arr_2d[0, :])  # Output: [1 2 3] (1st row, all columns)


Example for 3D array:

arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Accessing specific elements
print(arr_3d[0, 1, 0])  # Output: 3 (first block, second row, first column)

# Slicing 3D array
print(arr_3d[1, :, :])  # Output: [[5 6] [7 8]] (second block, all rows and columns)


4. Array Operations on Multidimensional Arrays:
NumPy allows you to perform element-wise operations on multidimensional arrays, just like it does with 1D arrays.

Example:

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Element-wise addition
print(arr_2d + 10)

# Element-wise multiplication
print(arr_2d * 2)

Output:

[[11 12 13]
 [14 15 16]]

[[ 2  4  6]
 [ 8 10 12]]

5. Reshaping Multidimensional Arrays:
You can change the shape of a multidimensional array using the reshape() method. This is particularly useful for preparing data for machine learning or performing matrix transformations.

#Example:
arr_1d = np.array([1, 2, 3, 4, 5, 6])

# Reshaping to 2D array
arr_2d = arr_1d.reshape(2, 3)
print(arr_2d)

# Reshaping to 3D array
arr_3d = arr_1d.reshape(2, 1, 3)
print(arr_3d)

#Output:

[[1 2 3]
 [4 5 6]]

[[[1 2 3]]
 [[4 5 6]]]

6. Broadcasting with Multidimensional Arrays:
NumPy also supports broadcasting, which is a way of applying operations on arrays of different shapes. When performing operations on arrays of different sizes, NumPy will broadcast the smaller array across the larger array to make them compatible.

#Example:
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_1d = np.array([10, 20, 30])

# Adding a 1D array to a 2D array (broadcasting happens)
print(arr_2d + arr_1d)

#Output:
[[11 22 33]
 [14 25 36]]

Here, the 1D array [10, 20, 30] is broadcast across the rows of the 2D array, adding the values element-wise.


Summary of NumPy’s Handling of Multidimensional Arrays:
Creation: Easily create multidimensional arrays using nested lists or np.reshape().

Shape and Dimensions: Check the shape (.shape) and number of dimensions (.ndim) of arrays.

Indexing and Slicing: Use indices and slicing to access and manipulate array elements.

Operations: Perform element-wise and matrix operations on multidimensional arrays.

Reshaping: Change the shape of an array with reshape().

Broadcasting: Perform operations on arrays of different shapes through broadcasting.

Q17- Explain the difference between apply() and map() in Pandas?
Ans- Both apply() and map() in Pandas are used to apply functions to data, but they are used in slightly different contexts and have distinct capabilities.

map() — Used for Element-Wise Operations on Series
Applies a function to each element in a Series (one-dimensional).
Typically used with simple operations or transformations.
Can take:
A function
A dictionary (for mapping values)
A Series (for element-wise mapping)

#Example:

import pandas as pd

s = pd.Series([1, 2, 3, 4])
s_mapped = s.map(lambda x: x * 2)
print(s_mapped)

#Output:
0    2
1    4
2    6
3    8
dtype: int64

apply() — Used for More Flexible Operations on Series or DataFrames
Can be used on both Series and DataFrames.
On a Series: behaves like map() but supports more complex logic.
On a DataFrame: can apply a function across rows (axis=1) or columns (axis=0).
More flexible for row/column-wise operations.

Example on a Series:
s = pd.Series([1, 2, 3])
s_applied = s.apply(lambda x: x ** 2)
print(s_applied)

Example on a DataFrame:
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Apply a function to each column
print(df.apply(sum, axis=0))

# Apply a function to each row
print(df.apply(lambda row: row['A'] + row['B'], axis=1))

Summary:
Use map() for simple, element-wise transformations on a Series.
Use apply() when you need more control or are working with DataFrames.

Q21- Why is NumPy’s array slicing faster than Python’s list slicing?
Ans- NumPy’s array slicing is faster than Python’s list slicing mainly because of how NumPy arrays are stored and processed in memory:

1. Contiguous Memory Layout
NumPy arrays are stored in contiguous blocks of memory, like C arrays. This allows fast access to elements and efficient operations, since the memory is compact and predictable.

2. Vectorized Operations
NumPy slicing returns views (not copies) of data where possible, meaning it avoids duplicating data. Plus, NumPy uses optimized, low-level C code under the hood for slicing and other array operations.

3. Typed Data
NumPy arrays hold elements of the same data type, so the underlying operations don’t need to check data types at each step, unlike Python lists which can contain mixed types.

4. Lower Overhead
Python lists are essentially arrays of pointers to objects, each of which could be stored anywhere in memory. This adds extra overhead to access and manipulate elements.

So, NumPy is not just faster in slicing—it's faster for most numerical computations for all these reasons.
